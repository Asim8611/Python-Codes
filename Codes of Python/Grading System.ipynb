```python
students = [
    ('Asim', '2022-CS-547', {'PF': 92, 'OOP': 88, 'DSA': 85}),
    ('Abdullah', '2022-CS-525', {'PF': 85, 'OOP': 90, 'DSA': 78}),
    ('Abdulahad', '2022-CS-549', {'PF': 60, 'OOP': 85, 'DSA': 82})
]
for student in students:
    name, studentID, grades = student
    total_grade = 0
    num_subj = len(grades)
    
    print(f"\nName: {name}, ID: {studentID}")
    
    for subject, score in grades.items():
        if score >= 90:
            grade = 'A'
            grade_point = 4.0
        elif score >= 80:
            grade = 'B'
            grade_point = 3.0
        elif score >= 70:
            grade = 'C'
            grade_point = 2.0
        elif score >= 60:
            grade = 'D'
            grade_point = 1.0
        else:
            grade = 'F'
            grade_point = 0.0
        
        total_grade += grade_point
        print(f"{subject}: {score} ({grade})")
    
    gpa = total_grade / num_subj
    print(f"GPA: {gpa:.2f}")
```

```python
students = [
    ('Asim', '2022-CS-547', {'PF': 92, 'OOP': 88, 'DSA': 85}),
    ('Abdullah', '2022-CS-525', {'PF': 85, 'OOP': 90, 'DSA': 78}),
    ('Abdulahad', '2022-CS-549', {'PF': 60, 'OOP': 85, 'DSA': 82})
]

highest_score = {'name': None, 'score': 0}
total_gpa_sum = 0
total_students = len(students)

for student in students:
    name, studentID, grades = student  
    total_grade = 0 
    num_subj = len(grades)  
    
    print(f"\nReport for {name} (ID: {studentID})")
    needs_improvement = False 
    excels_in_subjects = False  
    
    for subject, score in grades.items():
        if score >= 90:
            grade = 'A'
            grade_point = 4.0
            excels_in_subjects = True  
        elif score >= 80:
            grade = 'B'
            grade_point = 3.0
        elif score >= 70:
            grade = 'C'
            grade_point = 2.0
        elif score >= 60:
            grade = 'D'
            grade_point = 1.0
            needs_improvement = True  
        else:
            grade = 'F'
            grade_point = 0.0
            needs_improvement = True  
        
        total_grade += grade_point  
        print(f"{subject}: {score} ({grade})")
        
        if score > highest_score['score']:
            highest_score['name'] = name
            highest_score['score'] = score

    gpa = total_grade / num_subj
    print(f"GPA: {gpa:.2f}")
    total_gpa_sum += gpa  
    
    if excels_in_subjects:
        print("Performance: Great Performance.")
    if needs_improvement:
        print("Performance: Needs improvement in certain subjects.")

average_gpa = total_gpa_sum / total_students 
print("\nClass-Wide Performance Summary:")
print(f"Highest score: {highest_score['score']} by {highest_score['name']}")
print(f"Average GPA: {average_gpa:.2f}")
```

```python

students = {
    '2022-CS-547': {
        'name': 'Asim',
        'semesters': {
            'Semester 1': {'PF': 'A', 'OOP': 'B', 'DSA': 'B'},
            'Semester 2': {'TWAPS': 'A', 'TOA': 'B', 'DM': 'B'}
        }
    }
}

students['2022-CS-547']['semesters']['Semester 1']['OOP'] = 'A'

for semester, grades in students['2022-CS-547']['semesters'].items():
    print(f"{semester}: {grades}")

letter_to_gpa = {'A': 4.0, 'B': 3.0, 'C': 2.0, 'D': 1.0, 'F': 0.0}
total_grade_points = 0
total_subjects = 0

for grades in students['2022-CS-547']['semesters'].values():
    for grade in grades.values():
        total_grade_points += letter_to_gpa.get(grade, 0)
        total_subjects += 1

cumulative_gpa = total_grade_points / total_subjects
print(f"CGPA: {cumulative_gpa:.2f}")
```
